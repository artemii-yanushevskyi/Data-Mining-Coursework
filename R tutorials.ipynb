```r
# Load packages
library(ggplot2)
install.packages('GGally')
library(GGally)
```

```r
my.data <- read.csv("~/Downloads/reduced_data.csv", # path to the .csv file, modify if needed 
                    header = TRUE) # Does the .csv file contain a header row?
```

```r
summary(my.data) # Plot summary statistics for each attribute
```

```r
 # Get the class (type) of all attributes by (l)applying the function "class" # to all columns of our dataframe "my.data".
lapply(my.data, class)
```

```r
my.data$Store <- as.factor(my.data$Store) # Convert Store to factor and # overwrite it in my.data
summary(my.data)
```

```r
tmp <- paste(my.data$Date, "-01", sep = "") # Set all days to 01 
my.data$Date <- as.Date(tmp) # Overwrite my.data$Date
```

```r
ggpairs(my.data)
```

```r
mp1 <- ggplot(my.data,
    aes(x = Date, y = Sales, col = Store))

mp1 + geom_point() + geom_line() +
    theme(axis.text.x = element_text(
                                    angle = 90, # This just adjusts the x-axis text 
                                    hjust = 1, # Try removing/commenting and check
                                    size = 6) # the results. 
         ) 
```

```r
mp2 <- ggplot(my.data, aes(x = Store, y = Sales, fill = Store))
mp2 + 
  geom_boxplot(outlier.shape = NA) + # do not plot outliers, since I'll add the points later
  geom_point(alpha = 0.4,                            # point transparency
             size = 1,                               # point size
             position = position_jitter(width = 0.1),# add some jitter for clearer view 
             pch = 16)                               # point shape 
```

```r
mp2 + 
  geom_violin(width = 1.5) +
  #geom_boxplot(outlier.shape = NA, alpha = 0.2) + # uncomment and check what happens
  geom_point(alpha = 0.4, size = 1, 
             position = position_jitter(width = 0.05), pch = 16)
```

```r
mp3 <- ggplot(my.data, aes(x = Unemployment, y = Sales, colour = Fuel_Price))
mp3 + geom_point() + 
  scale_color_gradient(low = "#0088ff", high = "#ff5500") + # better color scale
  facet_grid(Store ~ .) # facet by store


mp4 <- ggplot(my.data, aes(x = Date, y = Sales, group = Store, 
                           colour = Unemployment))
mp4 + geom_line() + 
  scale_color_gradient(low = "#0088ff", high = "#ff5500") +
  theme(axis.text.x = element_text(angle = 90, hjust = 1, size = 6))
```
